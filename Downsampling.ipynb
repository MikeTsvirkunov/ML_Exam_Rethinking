```python
import numpy as np
import pandas as pd
import scipy
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import joblib
from imblearn.under_sampling import RandomUnderSampler
```

```python
data_with_sm_embendings = pd.read_csv("Data/smoted_data_with_sm_embendings.csv")
x_with_sm_embendings = data_with_sm_embendings.drop('State', axis=1)
y_with_sm_embendings = data_with_sm_embendings['State']
data_normed = pd.read_csv("Data/data_normed.csv")
x_normed = data_normed.drop('State', axis=1)
y_normed = data_normed['State']
```

```python
nomed = RandomUnderSampler(random_state=42)
x_res, y_res = nomed.fit_resample(x_normed, y_normed)
```

```python
randomundersampled_data_normed = pd.concat([x_res, y_res], axis=1)
randomundersampled_data_normed
```

```python
randomundersampled_data_normed.to_csv('Data/randomundersampled_data_normed.csv', index=False)
```

```python
sm = RandomUnderSampler(random_state=42)
x_res, y_res = sm.fit_resample(x_with_sm_embendings, y_with_sm_embendings)
```

```python
randomundersampled_data_with_sm_embendings = pd.concat([x_res, y_res], axis=1)
randomundersampled_data_with_sm_embendings
```
