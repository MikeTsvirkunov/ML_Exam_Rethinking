```python
import numpy as np
import scipy
import pandas as pd
import json
```

```python
data = pd.read_csv("Data/projects.csv")
```

```python
data_exessless = data.drop(['ID', 'Launched', 'Deadline', 'Name'], axis=1)
```

# Factorization

```python
data_categoris, data_categoris_labels = data_exessless['Category'].factorize()
data_subcategory, data_subcategory_labels = data_exessless['Subcategory'].factorize()
data_country, data_country_labels = data_exessless['Country'].factorize()
data_state, data_state_labels = data_exessless['State'].factorize()
data_factorized = data_exessless.copy(deep=True)
data_factorized['Category'] = data_categoris
data_factorized['Subcategory'] = data_subcategory
data_factorized['Country'] = data_country
data_factorized['State'] = data_state
data_factorized
```

```python
with open('Data/categoris_labels.json', 'w') as cl:
    json.dump(dict(enumerate(data_categoris_labels.to_list())), cl)
with open('Data/subcategory_labels.json', 'w') as cl:
    json.dump(dict(enumerate(data_subcategory_labels.to_list())), cl)
with open('Data/country_labels.json', 'w') as cl:
    json.dump(dict(enumerate(data_country_labels.to_list())), cl)
with open('Data/state_labels.json', 'w') as cl:
    json.dump(dict(enumerate(data_state_labels.to_list())), cl)
```

# Date split

```python
data_launched_split = data_factorized.copy(deep=True)
launched = pd.to_datetime(data['Launched'])
data_launched_split['Year_launched'] = launched.dt.year
data_launched_split['Month_launched'] = launched.dt.month
data_launched_split['Day_launched'] = launched.dt.day
data_launched_split['Hours_launched'] = launched.dt.hour
data_launched_split['Minutes_launched'] = launched.dt.minute
data_launched_split['Seconds_launched'] = launched.dt.second
data_launched_split

```

```python
data_deadline_split = data_launched_split.copy(deep=True)
deadline = pd.to_datetime(data['Deadline'])
data_deadline_split['Year_deadline'] = deadline.dt.year
data_deadline_split['Month_deadline'] = deadline.dt.month
data_deadline_split['Day_deadline'] = deadline.dt.day
data_deadline_split
```

# Resultes

```python
data_date_split = data_deadline_split.copy(deep=True)
data_date_split
```
