```python
import numpy as np
import pandas as pd
import spacy
import threading
nlp = spacy.load('en_core_web_md')
nlpsm = spacy.load('en_core_web_sm')
```

```python
names = pd.read_csv("Data/projects.csv")['Name'].tolist()
names0 = pd.read_csv("Data/names_0_100000.csv")
names1= pd.read_csv("Data/names_100000_200000.csv")
names2 = pd.read_csv("Data/names_200000_300000.csv")
names3 = pd.read_csv("Data/names_300000_374853.csv")
data = pd.read_csv("Data/data_normed.csv")
```

```python
nlp(names[0]).vector.shape
```

```python
names_vec = pd.concat([names0, names1, names2, names3])
names_vec
```

```python
data_names_embeded_added = pd.concat([data, names_vec.reset_index().drop('index', axis=1)], axis=1)
data_names_embeded_added
```

```python
data_names_embeded_added.to_csv('Data/data_names_embeded_md_added.csv', index=False)
```

```python
n = np.array(list(map(lambda a: nlpsm(a).vector, names)))
ns = pd.DataFrame(data=n, columns=list(range(n.shape[1])))
ns
```

```python
data_names_embeded_sm_added = pd.concat([data, ns.reset_index().drop('index', axis=1)], axis=1)
data_names_embeded_sm_added
```

```python
data_names_embeded_sm_added.to_csv('Data/data_names_embeded_sm_added.csv', index=False)
```
